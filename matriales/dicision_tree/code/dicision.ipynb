```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
datasets_path = r'D:/diploma AI/advanced_machine_learning_course_Amit/matriales/dicision_tree/code/Position_Salaries.csv'
datasets = pd.read_csv(datasets_path)
datasets.head()
```

```python
x = datasets.iloc[:,1:2].values
y = datasets.iloc[:,2].values
```

```python
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(x,y)
```

```python
value_to_predict = 6.5
y_pred = regressor.predict([[value_to_predict]])
print(f"predicted salary for level {value_to_predict}: {y_pred[0]}")
```

```python
# Visualizing the Decision Tree Regression results (higher resolution and smoother curve)
X_grid = np.arange(min(x), max(x), 0.01)  # Generate a fine grid of values
X_grid = X_grid.reshape((len(X_grid), 1))  # Reshape to 2D array
plt.scatter(x, y, color='red', label='Actual Data')
plt.plot(X_grid, regressor.predict(X_grid), color='blue', label='Regression Curve')
plt.title('Decision Tree Regression Results')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
