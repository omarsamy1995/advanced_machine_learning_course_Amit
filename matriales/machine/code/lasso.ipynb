```python
import pandas as pd 
import matplotlib.pyplot as plt
import os

dataset = pd.read_csv(r'D:/diploma AI/advanced_machine_learning_course_Amit/matriales/machine/code/datadata.txt')
dataset
```

```python
df = dataset.copy()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import Ridge 
l = 1 #lambda
ridge = Ridge(alpha=1)
ridge.fit(x,y)
```

```python
ridge_coef = ridge.coef_
ridge_coef
```

```python
ridge_intercept = ridge.intercept_
ridge_intercept
```

```python
x_test = [1,2,3,4]
y_pred = ridge.predict(pd.DataFrame(x_test))
y_pred
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(x_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(x,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

plt.text(x_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %ridge_coef)  + ' * X +'  +  str('%.2f' %ridge_intercept)    
                                                   +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('Ridge Regression')
plt.show()
```
